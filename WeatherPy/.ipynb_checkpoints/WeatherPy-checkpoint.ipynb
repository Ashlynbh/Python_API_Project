## Generate Cities List

```python
#Output data 
output_file = "../Data4/cities.csv"

# latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# List for holding co-ordinates 
cities = []
lat_lngs= []
found_cities = []

# random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each random combo
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # append unique cities into city list
    if city not in cities:
        cities.append(city)

#Create cities list 
city_count = len(cities)

print(city_count)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# create a params dict that will be updated with new city each iteration
params = {"key": weather_api_key}
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
query_url = f"{base_url}appid={weather_api_key}&units={units}&q="


# Creates lists for data to be stored 
lat = []
lon = []
found_cities = []
temp = []
humidity = []
cloudiness = []
wind_speed = []
date = []
country = []
set_count = 1
city_count = 0 

# Data loading is beginning 
print('--------------------------------')
print('Data Retrieval Beginning     ')
print('--------------------------------') 
         

for city in cities:     
    try:
        response = requests.get(query_url + city).json()
        lon.append(response["coord"]["lon"])
        lat.append(response["coord"]["lat"])
        temp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        cloudiness.append(response["clouds"]["all"])
        wind_speed.append(response["wind"]["speed"])
        date.append(response["dt"])
        country.append(response["sys"]["country"])

        if city_count >= 50:
            city_count = 1
            set_count += 1
            found_cities.append(city)
        else:
            city_count += 1
            found_cities.append(city)
        print(f"Processing Record {city_count} of Set {set_count} | {city}")

    except:
        print('City not found. Skipping...')
        pass
              
#Data Loading is complete 
print('--------------------------------')
print('Data Retrieval Complete      ')
print('--------------------------------') 

         
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a dictionary for establishing dataframe
cities_data = {
    "City":found_cities,
    "Latitude":lat,
    "Longitude":lon,
    "Max Temp":temp,
    "Humidity":humidity,
    "Cloudiness":cloudiness,
    "Wind Speed":wind_speed,
    "Country":country,
    "Date":date,
}
cities_df = pd.DataFrame(cities_data)

#Exports DataFrame as a csv file
cities_df.to_csv('output_file', index= False)

# Previews table/DataFrame
cities_df.head(100)
```

## Linear Regression

```python
# Dataframes for the different hemispheres - we know the equator is 0 
north_df = clean_city_data.loc[(clean_city_data['Latitude'] >= 0)]
south_df = clean_city_data.loc[(clean_city_data['Latitude'] < 0)]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_df['Latitude']
y_values = north_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,40),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')

print(f"The r-squared is: {rvalue**2}")

plt.show()
```

```python
x_values = south_df['Latitude']
y_values = south_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(-50,20),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')

print(f"The r-squared is: {rvalue**2}")

plt.show()
```
